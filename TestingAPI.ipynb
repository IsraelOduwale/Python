```python
import requests
response = requests.get("https://randomuser.me/api/")
print(response.status_code)
print(response.text)
print(response.reason)
```

```python
##Another simple example with a slightly larger output
response = requests.get("https://api.thedogapi.com/v1/breeds")
print(response.text)
```

## Querying data in API Example with inputs

```python
response = requests.get("https://randomuser.me/api/?gender=female")
print(response.text)
```

```python
response = requests.get("https://randomuser.me/api/?gender=male")
print(response.text)
```
